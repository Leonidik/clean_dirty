```python
import requests
```

```python
from os import listdir

from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PATH_to_clean = './data/clean/'
path_to_clean = []
for path in sorted(listdir(PATH_to_clean)):
    path_to_clean.append(PATH_to_clean + path)   
N_clean = len(path_to_clean) 
print(path_to_clean[0:1])

PATH_to_dirty = './data/dirty/'
path_to_dirty = []
for path in sorted(listdir(PATH_to_dirty)):
    path_to_dirty.append(PATH_to_dirty + path)
N_dirty = len(path_to_dirty) 
print(path_to_dirty[0:1])

path_to_files = path_to_clean + path_to_dirty
N_files       = N_clean + N_dirty
print()
print('Всего файлов с изображениями тарелок        : ', N_files)
print('Всего файлов с изображениями чистых тарелок : ', N_clean)
print('Всего файлов с изображениями грязных тарелок: ', N_dirty)
```

```python
path = 'http://176.107.241.37:8000/'
path = 'http://192.168.1.130:8000/'
path = 'http://127.0.0.1:8000/'
```

```python
# Получение токена / Регистрация и получение токена 
request = {
   'username': 'ivanov',
   'password': 'ivanov' }

response = requests.post(
    url = path + 'auth/',
    data=request)

print(response)
print('Text :',response.json()['text'])
print('token:',response.json()['token'])
```

```python
token = 'Token ' + response.json()['token']
token
```
