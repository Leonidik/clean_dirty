**PIL**

```python
from PIL import Image
```

```python
#img = Image.open('data/clean/009c.jpg')
img = Image.open(path)
print(type(img))
img
```

```python
from PIL import Image

# Read image
image_PIL = Image.open(path)
print(type(image_PIL))

# Output Images
#image_PIL.show()

# prints format of image
print(image_PIL.format)

# prints mode of image
print(image_PIL.mode)

# prints size of image: 
print(image_PIL.size)

image_PIL
```

```python
import numpy as np

# PIL => numpy
image_np = np.array(image_PIL) # np.shape: height x width x channel
print(type(image_np))
print(image_np.shape)

# nparray => PIL
image_PIL = Image.fromarray(image_np) # PIL.size: width x height
print(type(image_PIL))
print(image_PIL.size)
```
